```python
%matplotlib notebook
import pytesseract

import os
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image
from PIL import ImageOps
```

```python
# path and names
img_dir = '/home/ubuntu/data/Screenshots/'
img_name_list = sorted(os.listdir(img_dir))
example_name = img_name_list[-1]
img_name_list = img_name_list[:-1]
```

```python
pil_img = Image.open(os.path.join(img_dir, img_name_list[5]))
np_img = np.array(pil_img)
```

```python
crop_dict = dict()


# eliminations
# crop_eliminations = np_img[890:920, 131:157]
crop_eliminations = np_img[890:920, 132:157]
crop_dict['eliminations'] = crop_eliminations

# objective kills
crop_kills = np_img[890:920, 382:406]
crop_dict['kills'] = crop_kills


# objective time
crop_time = np_img[890:920, 625:700]
crop_dict['time'] = crop_time


# damage done
crop_damage = np_img[955:980, 130:205]
crop_dict['damage'] = crop_damage


# healing done
crop_healing = np_img[955:980, 380:405]
crop_dict['healing'] = crop_healing

# deaths
# crop_deaths = np_img[955:980, 630:655]
# crop_deaths = np_img[955:980, 630:652]
crop_deaths = np_img[955:980, 632:654]
crop_dict['deaths'] = crop_deaths

keys = ['eliminations', 'kills', 'time', 'damage', 'healing', 'deaths']

# crops = [crop_eliminations, crop_kills, crop_time, crop_damage, crop_healing, crop_deaths]
```

```python
plt.figure(figsize=(12, 2))
plt.imshow(np_img[800:1020, 0:800])
# plt.imshow(crop_dict['deaths'])
_ = plt.axis(False)
```

```python
for key in keys:
    ocr_result = pytesseract.image_to_string(crop_dict[key], config='--psm 7 --oem 0 -c tessedit_char_whitelist=0123456789,:')
    if ocr_result == '':
        crop = np.hstack([crop_dict[key], crop_dict[key], crop_dict[key]])
        ocr_result = pytesseract.image_to_string(crop, config='--psm 7 --oem 0 -c tessedit_char_whitelist=0123456789,:')
        if len(ocr_result) > 2:
            ocr_result = ocr_result[len(ocr_result)//3]
        else:
            ocr_result = '1'
    
    print('[{}] {}'.format(key, ocr_result))
```

```python
def box_lines(box, img_y_size):
    x = [box[0], box[0], box[2], box[2], box[0]]
    y = [img_y_size-box[1], img_y_size-box[3], img_y_size-box[3], img_y_size-box[1], img_y_size-box[1]]
    return x, y


cropped_img = np_img[800:1020, 0:800]

lines = pytesseract.image_to_boxes(cropped_img).split('\n')

boxes = []
chars = []

for line in lines:
    box_elements = line.split(' ')
    chars.append(box_elements[0])
    boxes.append([int(x) for x in box_elements[1:-1]])
    
# for char, box in zip(chars, boxes):
#     print(char, box)
    
    
fig = plt.figure(figsize=(10, 3))
ax = fig.add_subplot(111)
plt.show()
ax.imshow(cropped_img)
# ax.imshow(ImageOps.flip(cropped_img))

ImageOps

for box in boxes:
    x, y = box_lines(box, cropped_img.shape[1])
    plt.plot(x, y, 'y-')
    
# plt.axis('off')

```
